# Exericio de Regressão com TF - Avaliando preço de casas



California Housing Data

This data set contains information about all the block groups in California from the 1990 Census. In this sample a block group on average includes 1425.5 individuals living in a geographically compact area. 

The task is to aproximate the median house value of each block from the values of the rest of the variables. 

 It has been obtained from the LIACC repository. The original page where the data set can be found is: http://www.liaad.up.pt/~ltorgo/Regression/DataSets.html.
 

The Features:
 
* housingMedianAge: continuous. 
* totalRooms: continuous. 
* totalBedrooms: continuous. 
* population: continuous. 
* households: continuous. 
* medianIncome: continuous. 
* medianHouseValue: continuous. 

## The Data

** Import the cal_housing.csv file with pandas. Separate it into a training (70%) and testing set(30%).**

```python
import pandas as pd
```

```python
# data.frame
housing = pd.read_csv('cal_housing_clean.csv')
```

```python
housing.head()
```

```python
# pandas.describe: Generates descriptive statistics that 
# summarize the central tendency, dispersion and shape 
# of a dataset’s distribution, excluding NaN values.

# pandas.tranpose: converte index em colunas, e vice-versa
housing.describe().transpose()

# Usando os dois juntos obetmos  (com melhor visualizaçâo pelo tranpose)
# uma vizão estatistica de nosos dados
```

```python
# x_data ==> colunas de features
x_data = housing.drop(['medianHouseValue'], axis=1)
```

```python
# y_data => target
y_val = housing['medianHouseValue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data,y_val,
    test_size=0.3,random_state=101)
```

### Scale the Feature Data

** Use sklearn preprocessing to create a MinMaxScaler for the feature data. Fit this scaler only to the training data. Then use it to transform X_test and X_train. Then use the scaled X_test and X_train along with pd.Dataframe to re-create two dataframes of scaled data.**

# Leia
O Scaler faz algo muito importante.

Por exemplo: imagine que você tem uma rede neural e que nas suas fetaures tinha a idade e o preço de uma casa. A idade varia entre cerca de 10 à 90, enquanto que o preço de uma casa varia de 20.000 à 1.000.000.

Perceba que o intervalo de dados (o Max e Min dessas features) são bem diferentes. Como a rede neural faz somatorias multiplicando pesos, entâo, haverá impacto, ele em si nâo vai perceber que por exemplo, uma idade de 80 é grande, sendo que o preço minimo de uma casa é 20.000

Então **Para isso formatamos esses dados**. O escolador MinMax jogar os valores da features para [0,1] de acordo com o seu valor em relaçâo à max mim.

Dessa forma, as features de idade e preço ficarao num mesmo intervalo  ([0,1]), e assim, nâo haverá muita

Para isso usamos o MinMaxScaler. Qualquer coisa, consulte a lib  na doc

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

## Como era antes o X_train

```python
X_train.head()
```

```python
scaler.fit(X_train)
# Vmaos fitar soemtne os dados de treinamento
```

```python
# O scaler gera um numpy array, aqui, convertemos isso para um dataFrame
X_train = pd.DataFrame(
    data=scaler.transform(X_train),
    columns = X_train.columns,index=X_train.index)
```

## Como é agora o X_train depois do Scaler


```python
X_train.head()
```

```python
# Aqui, vamos converter os dados de test. Na verdade, todo dado que entar
# agora deverá ser convertido com esse scaler, para que seus valores fique 
# entre [0,1]
X_test = pd.DataFrame(
    data=scaler.transform(X_test),
    columns = X_test.columns,index=X_test.index)
```

### Create Feature Columns

** Create the necessary tf.feature_column objects for the estimator. They should all be trated as continuous numeric_columns. **

```python
housing.columns
```

```python
import tensorflow as tf
```

## As Coluans de Features para o tensorFlow tem que ser criadas uma a uma

```python
age = tf.feature_column.numeric_column('housingMedianAge')
rooms = tf.feature_column.numeric_column('totalRooms')
bedrooms = tf.feature_column.numeric_column('totalBedrooms')
pop = tf.feature_column.numeric_column('population')
households = tf.feature_column.numeric_column('households')
income = tf.feature_column.numeric_column('medianIncome')
```

```python
# Aqui, jautamos todas as coluans criads pelo TF ANTES
feat_cols = [ age, rooms, bedrooms, pop, households, income]
```

# Input Function
** Create the input function for the estimator object. (play around with batch_size and num_epochs)**

```python
input_func = tf.estimator.inputs.pandas_input_fn(
    x = X_train, y = y_train , batch_size = 10,
    num_epochs = 1000, shuffle = True)
```

** Create the estimator model. Use a DNNRegressor. Play around with the hidden units! **

## Vai ser uma DNN de 3 camadas, 6 neuronios em cada

```python

model = tf.estimator.DNNRegressor(
    hidden_units=[6,6,6], feature_columns=feat_cols)
```

# TRAINANDO O MODELO
##### ** Train the model for ~1,000 steps. (Later come back to this and train it for more and check for improvement) **

```python
model.train(input_fn=input_func, steps=25000)
```

# PREDIZENDO AGORA OS DADOS DE TEST
** Create a prediction input function and then use the .predict method off your estimator model to create a list or predictions on your test data. **

```python
# Quando se prediz: shuffle é false e nâo há y
predict_input_func = tf.estimator.inputs.pandas_input_fn(
      x = X_test,
      batch_size = 10,
      num_epochs = 1,
      shuffle = False)
# Craidmos a funçâo que prediz
```

```python
# Agora camos predizer
pred_gen = model.predict(predict_input_func)
```

```python
predictions = list(pred_gen)
```

```python
# Vendo a prediçâo: (Formato do TF)
predictions
```

** Calculate the RMSE. Do this manually or use [sklearn.metrics](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html) **

```python
final_preds = []
for pred in predictions:
    final_preds.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Erro da media ao quadrado: erra cerca de 97.00
mean_squared_error(y_test,final_preds)**0.5
```
