#### 6 - Groupby
O método groupby permite agrupar linhas de dados em conjunto e chamar funções agregadas. Voce agrupa por alguma oluna e entao aplica operaçoes nas outras
+ Faz GROUP BY como no SQL, e da mesma forma, só serve para alicar operaçoes sobre esse grupo
    + sum() = soma
    + describe() = retorna varios metodos aplicados (muito bom
    + mean() = média 
    + min()
    + max()
    + transpose() = inverte linha x coluna da tabela (Fornece uma outra visao de dados)

```python
# Cria um DataFrame
data = {'Empresa':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Nome':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Venda':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
df
```

```python
## Aplicando GroupBy
df.groupby('Empresa').sum()
```

```python
## Perceba que ele nao esta operand sobre nome, por que percebe que, como em Nome é string, nao vai executar esas operçaoes
df.groupby('Empresa').describe()
```

```python
###  .trasnpose
df.groupby('Empresa').describe().transpose()
```

#### 7 - Junçao de DataFrames
+ Mesclar, Juntar, e Concatenar
+ There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating. In this lecture we will discuss these 3 methods with examples.
+ Existem três maneiras principais de combinar os DataFrames: mesclando, juntando e concatenando (merge, join e concat). Nesta palestra, discutiremos esses 3 métodos com exemplos.



```python
## Dados para mesclar
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

###### Concatenação `concat()`
+ Concatenação basicamente cola DataFrames. Tenha em mente que as dimensões devem corresponder ao longo do eixo que você está concatenando. Você pode usar pd.concat e passar uma lista de DataFrames para concatenar juntos:
+ é imporatnte que o eixo que voce esta unindo os Data Frames tenham o mesmo tamanho
+ As unios sao feitas em coluna com mesmo nome
+ axis por efault é 0 (concatena linha (vertical)) e axis = 1 seria para colunas (cresce horizonal)
+ Planeje os indices e como voce quer que saia

```python
## juntou as linhas
pd.concat([df1, df2, df3])
```

```python
## juntou coluans, mas da problemas pois os indices de linhas sao diferentes
pd.concat([df1, df2, df3], axis = 1)
```

###### Mesclar `.merge()`
A função mesclar permite que você mescle os quadros de dados juntos usando uma lógica semelhante à mesclagem de tabelas SQL juntas. Por exemplo:
+ É O INNER JOIN DO SQL
+ Junta por uma coluna que tenham valore sigauis na duas, nao é o concatenar pois essa coluna fica la (desde que as duas a tenham)
+ @how =  inner/left/right/outer
+ @on  = coluna para mesclar as duas (elas devem ter essa coluna com mesmos dados pra fazer isso corretamente)


```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
## junta pela key
pd.merge(esquerda,direita,how='inner',on='key')
```

###### Juntar `.join()`
Juntar é um método conveniente para combinar as colunas de dois DataFrames indexados potencialmente diferentes em um único resultado DataFrame.
+ Tentar juntar Dad Frames em que parte dos dados batem
+ Diferente dos outros métodos, é um método do dataFrame
+ O que mais o diferencia dos outros é que ele faz a junçâo  TOMANDO COMO BASE O DATA-FRAME QUE EXECUTOU e tentan adpata o do aprametro para si
+ usar @how = 'outer' eh bem útil pra juntar tudo, depois trata NaN
    + Jutna todos os dados, e mantem os que sao iguai em linha e coluna

```python
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
esquerda
```

```python
direita
```

```python
## concatena linha e coluna do dataFrame do param@direita no da esquerda
### como dataFram@esquerda nao tem a linha K3, ela nao vai
### como dataFram@direita  nao tem a linha K1 que tem no esquerda (que é a base), vai receber NaN
esquerda.join(direita)
```

```python
### Muito utilizado para pegar tudo
## Jutna todos os dados, e mantem os que sao iguai em linha e coluna
esquerda.join(direita, how='outer')
```

**`.pivot_trable`** : Reestrutura os dados para ter uma outra pespequitiva. **É muito interressante**, pois voce torna colunas em dados e indixe
